### Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Read the data

```python
path = "./lalonde.csv"
df = pd.read_csv(path)
df.head()
```

## Question 1

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True, figsize=(8,3))
sns.distplot(df[df.treat == 1]['re78'], ax=ax1)
sns.distplot(df[df.treat == 0]['re78'], ax=ax2)
```

## Question 2

```python
features = set(df).difference(set(['id', 'treat', 're78']))

for feature in features:
    fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True, figsize=(8,3))
    sns.distplot(df[df.treat == 1][feature], ax=ax1)
    sns.distplot(df[df.treat == 0][feature], ax=ax2)
```
